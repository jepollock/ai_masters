```python
import pandas as pd
import numpy as np
import umap.umap_ as umap

umap_manifold = umap.UMAP()

# Load the data
# treat binary as categorical data
# drop ID
# convert categorical fields to one-hot.
# Add "unknown" to every categorical field for missing data.
# convert all categorical data to one-hot encoding.

from sklearn.manifold import MDS

# This example script simply outputs the node's input table.
data = np.array([[5.8, 2.8], [6.0, 2.2]])
input_table_1 = pd.DataFrame.from_records(data)

n_components = 5
dissimilarity = 'euclidean'

mds_manifold = MDS(n_components=n_components, dissimilarity=dissimilarity)
mds_manifold.fit_transform(input_table_1)
```
